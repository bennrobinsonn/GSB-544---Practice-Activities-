Set up

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Use other vars to predict bill_depth_mm
features = ['species', 'island', 'sex',
            'bill_length_mm', 'flipper_length_mm',
            'body_mass_g', 'year']

penguins2 = penguins.dropna(subset=features + ['bill_depth_mm'])
X = penguins2[features]
y = penguins2['bill_depth_mm']

cat_cols = ['species', 'island', 'sex']
num_cols = ['bill_length_mm', 'flipper_length_mm', 'body_mass_g', 'year']

preprocess = ColumnTransformer(
    [
        ("cat_ohe",
         OneHotEncoder(handle_unknown="ignore", sparse_output=False),
         cat_cols),
        ("num_scale",
         StandardScaler(),
         num_cols),
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=123
)
```

LR Model


```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_pipe = Pipeline(
    [
        ("preprocess", preprocess),
        ("model", LinearRegression())
    ]
)

lin_pipe.fit(X_train, y_train)

y_train_hat = lin_pipe.predict(X_train)
y_test_hat  = lin_pipe.predict(X_test)

print("Linear Reg – train MSE:", mean_squared_error(y_train, y_train_hat))
print("Linear Reg – test MSE: ", mean_squared_error(y_test, y_test_hat))
```

Knn with 3

```python
from sklearn.neighbors import KNeighborsRegressor

knn3_pipe = Pipeline(
    [
        ("preprocess", preprocess),
        ("model", KNeighborsRegressor(n_neighbors=3))
    ]
)

knn3_pipe.fit(X_train, y_train)

y_train_hat = knn3_pipe.predict(X_train)
y_test_hat  = knn3_pipe.predict(X_test)

print("kNN (k=3) – train MSE:", mean_squared_error(y_train, y_train_hat))
print("kNN (k=3) – test MSE: ", mean_squared_error(y_test, y_test_hat))
```

Knn wiht 15

```python
knn15_pipe = Pipeline(
    [
        ("preprocess", preprocess),
        ("model", KNeighborsRegressor(n_neighbors=15))
    ]
)

knn15_pipe.fit(X_train, y_train)

y_train_hat = knn15_pipe.predict(X_train)
y_test_hat  = knn15_pipe.predict(X_test)

print("kNN (k=15) – train MSE:", mean_squared_error(y_train, y_train_hat))
print("kNN (k=15) – test MSE: ", mean_squared_error(y_test, y_test_hat))
```

Decision Tre

```python
from sklearn.tree import DecisionTreeRegressor

tree_pipe = Pipeline(
    [
        ("preprocess", preprocess),
        ("model", DecisionTreeRegressor(random_state=123))
    ]
)

tree_pipe.fit(X_train, y_train)

y_train_hat = tree_pipe.predict(X_train)
y_test_hat  = tree_pipe.predict(X_test)

print("Tree – train MSE:", mean_squared_error(y_train, y_train_hat))
print("Tree – test MSE: ", mean_squared_error(y_test, y_test_hat))
```
